```python
import matplotlib.pyplot as plt

# Data
labels = list(range(10, 151))
accuracy = [7.34] * len(labels)  # 7.34% accuracy for all labels

# Plotting
plt.plot(labels, accuracy, label='GPT-3.5', marker='o')
plt.title('Accuracy of GPT-3.5 Model')
plt.xlabel('Number of Labels')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True)
plt.show()
```
