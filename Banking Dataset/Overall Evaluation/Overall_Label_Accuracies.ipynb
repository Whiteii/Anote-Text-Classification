```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
bert_label_accuracies = pd.read_csv('Bert_Banking_Data_Label_Accuracies.csv')
rnn_label_accuracies = pd.read_csv('RNN_banking_labels_metrics_result.csv')
ft_gpt_label_accuracies = pd.read_csv('ft_banking_labels_metrics_results.csv')
```

#  Accuracy, Precision, Recall Comparison Plot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(bert_label_accuracies['Labels'], bert_label_accuracies['accuracy'], label='BERT', marker='o', linestyle='-', color='blue', linewidth=2)
plt.plot(rnn_label_accuracies['Label'], rnn_label_accuracies['Accuracy'], label='RNN', marker='s', linestyle='--', color='green', linewidth=2)
plt.plot(ft_gpt_label_accuracies['Label'], ft_gpt_label_accuracies['accuracy'], label='FT-GPT', marker='^', linestyle='-', color='orange', linewidth=2)
plt.xlabel('Label')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

```python
plt.figure(figsize=(15, 5))

# Plot Precision
plt.subplot(1, 2, 1)
plt.plot(bert_label_accuracies['Labels'], bert_label_accuracies['precision'], label='BERT', marker='o', linestyle='-', color='blue', linewidth=2)
plt.plot(rnn_label_accuracies['Label'], rnn_label_accuracies['Precision'], label='RNN', marker='s', linestyle='--', color='green', linewidth=2)
plt.plot(ft_gpt_label_accuracies['Label'], ft_gpt_label_accuracies['precision'], label='FT-GPT', marker='^', linestyle='-', color='orange', linewidth=2)

plt.xlabel('Label')
plt.ylabel('Precision')
plt.title('Precision Comparison')
plt.legend()
```

```python
# Plot Recall
plt.subplot(1, 2, 2)
plt.plot(bert_label_accuracies['Labels'], bert_label_accuracies['recall'], label='BERT', marker='o', linestyle='-', color='blue', linewidth=2)
plt.plot(rnn_label_accuracies['Label'], rnn_label_accuracies['Recall'], label='RNN', marker='s', linestyle='--', color='green', linewidth=2)
plt.plot(ft_gpt_label_accuracies['Label'], ft_gpt_label_accuracies['recall'], label='FT-GPT', marker='^', linestyle='-', color='orange', linewidth=2)

plt.xlabel('Label')
plt.ylabel('Recall')
plt.title('Recall Comparison')
plt.legend()

plt.tight_layout()  
plt.show()
```
